```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("employee_data.csv")

# Plot the distribution of productivity
sns.distplot(df["targeted_productivity"])
plt.show()

# Plot the relationship between productivity and smv
sns.scatterplot(x="smv", y="targeted_productivity", data=df)
plt.show()

# Plot the relationship between productivity and wip
sns.scatterplot(x="wip", y="targeted_productivity", data=df)
plt.show()

# Plot the relationship between productivity and over_time
sns.scatterplot(x="over_time", y="targeted_productivity", data=df)
plt.show()

# Plot the relationship between productivity and incentive
sns.scatterplot(x="incentive", y="targeted_productivity", data=df)
plt.show()

# Plot the relationship between productivity and idle_time
sns.scatterplot(x="idle_time", y="targeted_productivity", data=df)
plt.show()

# Plot the relationship between productivity and no_of_workers
sns.scatterplot(x="no_of_workers", y="targeted_productivity", data=df)
plt.show()
```
