1. Use the following information to create a line plot:

    First, define a variable $\theta$ over 100 equally spaced intervals between 0 and 2$\pi$. 

    Then, generate data for variable $x$ according to the following equation:

    $ x = 100 \cdot sin^3(\theta) $

    and generate data for variable $y$ according to the following:

    $ y = 10 \cdot cos(\theta) - 5 \cdot cos(2\theta) - 2 \cdot cos(3\theta) - cos(4\theta) $

    Using the Matlab-style interface, plot the resulting data as a line chart.

    Give the chart an appropriate title and the line an appropriate colour.

```python
#import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

#define theta
theta = np.linspace(0,2*np.pi,100)

#first equation
x = 100*np.sin(theta)**3
#second equation
y = (10*np.cos(theta)) - (5*np.cos(2*theta)) - (2*np.cos(3*theta)) - (np.cos(4*theta))
#give the plot a title
plt.title('HEART')
#plot and assign plot color
plt.plot(x,y,color='pink')




```

2. From the `sklearn.datasets` package import the `load_wine` functionality to load a multivariate data set showing the results of a chemical analysis of several wines grown in the same region of Italy but originating from 3 different cultivars.

    The following documentation gives an overview of the functionality and what is returned: https://scikit-learn.org/stable/modules/generated/sklearn.datasets.load_wine.html

    Load in the data using the appropriate method. Via your own exploration of the data, plot, using the matlab-style interface, a scatter chart of 'alcohol' on the x-axis against 'color_intensity' on the y-axis. Using a 'plasma' colour map, use the value of 'hue' as the colour value. There may be some overlapping points, so set the transparency value to 0.5. Finally, set the size according to the value of the 'flavanoids' scaled by a factor of 20.

    Ensure the chart axes are labelled appropriately and a colour bar is added to the figure with an appropriate label too.

```python
from sklearn.datasets import load_wine
import pandas as pd
import numpy as np
import seaborn as sns


#load the dataset
wine = load_wine()

#Conver to pandas dataframe
data=pd.DataFrame(data=np.c_[wine['data'],wine['target']],columns=wine['feature_names']+['target'])

#plot scatter plot of the data
sns.pairplot(data,x_vars=["alcohol"],y_vars=["color_intensity"], hue='target',palette='plasma',height=5,
             aspect=1,kind='scatter',plot_kws=dict(alpha=0.5, s= 20*data['flavanoids']))





```

3. Define a function that takes two arrays and returns the cosine of the values of the arrays multiplied by one another ($cos(x.y)$). Using this function, pass in two arrays of evenly spaced numbers from -10 to 10 over 50 intervals. From these arrays, calculate the resulting meshgrid from the two arrays. Finally, create a contour plot using these grids and set the colour map to 'turbo', labelling the axes appropriately and including a colour bar, also with a label.

```python
import numpy as np

#define function
def cosine(x,y):
    return np.cos(x*y)

#define arrays
x = np.linspace(-10,10,50)
y = np.linspace(-10,10,50)

#calculate meshgrid
X,Y = np.meshgrid(x,y)

#create contour plot
plt.contourf(X,Y,cosine(X,Y),cmap='turbo')
#add colorbar
plt.colorbar(label='cosine')
#label axes
plt.xlabel('x')
plt.ylabel('y')





```
